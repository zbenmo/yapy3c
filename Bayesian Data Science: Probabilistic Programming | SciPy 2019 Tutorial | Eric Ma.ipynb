```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)

x = np.random.binomial(10, 0.5, 10_000)
```

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n + 1) / n

    return x, y
```

```python
x_flips, y_flips = ecdf(x)
```

```python
plt.plot(x_flips, y_flips, marker='.', linestyle='none');
```
